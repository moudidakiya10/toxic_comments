```python
dataset=tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset=dataset.cache()
dataset=dataset.shuffle(160000)
dataset=dataset.batch(16)
dataset=dataset.prefetch(8) #helps prevent bottlenecks
#MACSHBAP-map,cache,sheuffle,batch,prefetch from_tensor_slices,list_file
```

```python
train=dataset.take(int(len(dataset)*.7))
val=dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test=dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

### Create sequential model:

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import LSTM ,Dropout,Bidirectional,Dense,Embedding
```

```python
model= Sequential()
model.add(Embedding(NB_W+1,32))
model.add(Bidirectional(LSTM(32,activation='tanh')))
model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(6,activation='sigmoid'))
```

```python
model.compile(loss='BinaryCrossentropy',optimizer='Adam')
```

```python
model.summary()
```

```python
histoty=model.fit(train,epochs=1,validation_data=val)
```

### Evaluating the model

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre=Precision()
```

```python
re=Recall()
acc=CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    x_true,y_true=batch
    y_hat=model.predict(x_true)
    
    y_true=y_true.flatten()
    y_hat=y_hat.flatten()
    
    pre.update_state(y_true,y_hat)
    re.update_state(y_true,y_hat)
    acc.update_state(y_true,y_hat)
    
```

```python
print(f'Precision:{pre.result().numpy()},Recall:{re.result().numpy()},Accuracy:{acc.result().numpy()}')
```
